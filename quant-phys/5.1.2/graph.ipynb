```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Загрузка данных
data_df = pd.read_csv('data.csv')
mistakes_df = pd.read_csv('mistakes.csv')

# Переименование столбцов для удобства
data_df.columns = ['X', 'Y']
mistakes_df.columns = ['X_err', 'Y_err']

# Рассчет линейной трендовой линии
slope, intercept = np.polyfit(data_df['X'], data_df['Y'], 1)
trendline = slope * data_df['X'] + intercept

# Построение графика с погрешностями и линией тренда
plt.figure(figsize=(8, 6))
plt.errorbar(data_df['X'], data_df['Y'], xerr=mistakes_df['X_err'], yerr=mistakes_df['Y_err'], fmt='o', ecolor='red', capsize=5, label='$\dfrac{1}{N(\N{greek small letter theta})}$')
plt.plot(data_df['X'], trendline, color='blue', label=f'Trendline: y = {slope:.4f}x + {intercept:.4f}')

# Настройка осей и легенды
plt.xlabel('$1-cos \N{greek small letter theta}$')
plt.ylabel('$\dfrac{1}{N(\N{greek small letter theta})}$')
plt.grid()
plt.legend()

# Отображение сетки и графика
plt.savefig('plot.jpg', format='jpg')
plt.grid(True)
plt.show()
```
